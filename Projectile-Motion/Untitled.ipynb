```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np

x_data = [1, 2, 3, 4, 5]
y_data = [2, 3, 1, 5, 4]

fig, ax = plt.subplots()
ax.set_xlim(min(x_data) - 1, max(x_data) + 1)
ax.set_ylim(min(y_data) - 1, max(y_data) + 1)

line, = ax.plot([], [], 'b-')
point, = ax.plot([], [], 'ro')

def init():
    line.set_data([], [])
    point.set_data([], [])
    return line, point

def update(frame):
    line.set_data(x_data[:frame+1], y_data[:frame+1])
    point.set_data([x_data[frame]], [y_data[frame]])
    return line, point

ani = animation.FuncAnimation(fig, update, frames=len(x_data),
                              init_func=init, blit=True, interval=500, repeat=False)

HTML(ani.to_jshtml())
```

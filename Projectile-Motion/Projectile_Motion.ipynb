Projectile Motion With Drag

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
import numpy as np
```

Creating definitions:

```python
def EulersCalc(x, y, vx, vy, dt, time, B_over_m):
    '''
    Use Euler's method to calculate speed and velocity with drag 
    '''
    i=1
    while y[i-1] >= 0:

        x = np.append(x,x[i-1]+vx[i-1]*dt)
        y = np.append(y,y[i-1]+vy[i-1]*dt)
        v_tot = (vx[i-1]**2 + vy[i-1]**2)**(1/2)
        vx = np.append(vx,vx[i-1] + (B_over_m*v_tot*vx[i-1])*dt)
        vy = np.append(vy,vy[i-1] + (g + B_over_m*v_tot*vy[i-1])*dt)              
        time=np.append(time,time[i-1]+dt)
        
        i=i+1
    return(x,y,vx,vy,time,i-1)
```

```python
def EulersTimeCalc(x, y, vx, vy, dt, tme, B_over_m):
    '''
    Use Euler's method to calculate speed and velocity with drag, returns only the final found time.
    '''
    i=1
    while y >= 0:

        v_tot = (vx**2 + vy**2)**(1/2)
        vx += (B_over_m*v_tot*vx)*dt
        vy += (g + B_over_m*v_tot*vy)*dt  
        x += vx*dt
        y += vy*dt
        tme += dt
        
        i=i+1
    return(tme)
```

```python
#This is not necessary, but I'm thinking of converting this into finding the time, That said, I could just lower the time step for time to be 
#as accurate as I want it to be sooooooo...
def range(x,y,n):
    """
    This function deteremined the range of the projectile
    """
    r=-y[n-1]/y[n]
    distance=(x[n-1]+r*x[n])/(r+1)
    return distance
```

```python
def plot(d1, d2, t1, t2, label1,label2, color1, color2): #arry,array,array,array,str,str,str,str
    '''
    this function is to plot the distance vs time for two variables 
    '''
    plt.plot(t1,d1,label = label1,color = color1)
    plt.plot(t2,d2,label = label2,color = color2)
    plt.ylabel('distance (m)')
    plt.xlabel('time (s)')
    plt.title('Projectile Distance Over Time')
    plt.legend()
    plt.ylim([0,d1[0]+10])
    plt.grid(True)
    
    plt.show()    
```

Starting main body of code:

```python

'''
Initialize and define variables:
drag coefficient: C
time step: dt
density: dens
area: A
mass: m
height: h
starts arrays for x,y,x_drag,y_drag for both dropped and launched
'''

C = 0.5
g = -9.8 
dens = float(input("density of the medium:"))
A = float(input('Object area in meters squared:'))
m = float(input('Object mass in kilograms:'))
dt = float(input('Time step in seconds:'))
h = float(input('Starting height in meters:'))
speed = float(input('initial launch speed in meters per second:'))
                  
B_over_m = -((1/2)*C*dens*A)/m  

'''            
Starting two sets of arrays:
'''
    
#dropped object arrays:
xd = np.array([0])
yd = np.array([h])
vxd = np.array([0])
vyd = np.array ([0])
time_d = np.array([0])


#Launched objects array:
xl = np.array([0])
yl = np.array([h])
vxl = np.array([speed])
vyl = np.array([0])
time_l = np.array([0])
```

```python
'''
setting up arrays to graph from user inputted numbers
'''
xd, yd, vxd, vyd, time_d,nd = EulersCalc(xd,yd,vxd,vyd,dt,time_d,B_over_m)
xl,yl,vxl,vyl,time_l,nl = EulersCalc(xl,yl,vxl,vyl,dt,time_l,B_over_m)
plot(yd,yl,time_d,time_l,'Distance Dropped','Distance Launched','blue','red')
```

```python
'''
plotting arrays from user inputted numbers
'''
plt.plot(time_d,yd,label = 'Dropped Bullet',color = 'blue')
plt.plot(time_l,yl,label = 'Launched Bullet',color = 'red')
plt.xlabel('time (s)')
plt.ylabel('height (m)')
plt.title('Bullet Heigh vs Time')
plt.legend()
plt.grid(True)
#plt.savefig("bullet drop")

#plt.ylim([0,1550])

plt.show()    
```

```python
'''
Creating group of all the constnt values
'''
g = -9.81
h = 10000
C = 0.5
dens = 1.225
A = 0.1
m = 100
dt = 0.01
speed = 10                  
B_over_m = -((1/2)*C*dens*A)/m  

```

```python
''' 
Code to determine time difference between dropped and launched times to hit the ground
with a change in velocity
'''
time_diff = np.array([])
Speed = np.arange(10, 10000,10)

td = EulersTimeCalc(0, 10000, 0, 0, 0.01, 0, B_over_m)
for vx in Speed:
    local_td = td
    local_B_over_m = B_over_m
    
    tl = EulersTimeCalc(0, 10000, vx, 0, 0.01, 0, local_B_over_m)
    time_diff = np.append(time_diff, tl-local_td)



plt.plot(Speed, time_diff)
plt.xlabel('Initial Launch speed')
plt.ylabel('Time Difference (Launched minus Dropped)')
plt.title('Comparing Time to Hit the Ground From 10,000 Meters Different Launch Velocities')
plt.grid(True)

plt.show()  
```

```python
''' 
Code to determine time difference between dropped and launched times to hit the ground
with a change in mass
'''
time_diff_m = np.array([])
Mass = np.arange(20, 2000,10)


for x in Mass:

    local_B_over_m = -((1/2)*0.5*1.225*0.1)/x  

    t_d = EulersTimeCalc(0, 10000, 0, 0, 0.001, 0, local_B_over_m)
    t_l = EulersTimeCalc(0, 10000, 100, 0, 0.001, 0, local_B_over_m)
    time_diff_m = np.append(time_diff_m, t_l-t_d)



plt.plot(Mass, time_diff_m)
plt.xlabel('Mass')
plt.ylabel('Time Difference (Launched minus Dropped)')
plt.title('Comparing Time to Hit the Ground From 10,000 Meters Different Masses')
plt.grid(True)

plt.show()  
```

```python
''' 
Code to determine time difference between dropped and launched times to hit the ground
with a change in height
'''
time_diff_h = np.array([])
Height = np.arange(10, 10000, 10)


for H in Height:

    local_B_over_mh = -((1/2)*0.5*1.225*0.1)/100  

    t_dh = EulersTimeCalc(0, H, 0, 0, 0.001, 0, local_B_over_mh)
    t_lh = EulersTimeCalc(0, H, 100, 0, 0.001, 0, local_B_over_mh)
    time_diff_h = np.append(time_diff_h, t_lh-t_dh)



plt.plot(Height, time_diff_h)
plt.xlabel('Height')
plt.ylabel('Time Difference (Launched minus Dropped)')
plt.title('Comparing Time to Hit the Ground From different Heights')
plt.grid(True)

plt.show()  
```

```python
plt.plot(Speed, time_diff)
plt.xlabel('Launch Velocity (m/s)')
plt.ylabel('Time Difference Launched Minus Dropped (s)')
plt.title('Drag and Launch Speed Dependence')
plt.grid(True)
plt.savefig('Drag and Launch Speed Dependence')
#plt.show()

plt.plot(Height, time_diff_h)
plt.xlabel('Lunch Height (m)')
plt.ylabel('Time Difference Launched Minus Dropped (s)')
plt.title('Drag and Height Dependence')
plt.grid(True)
plt.savefig('Drag and Height Dependence')
#plt.show() 

plt.plot(Mass, time_diff_m)
plt.xlabel('Mass (kg)')
plt.ylabel('Time Difference Launched Minus Dropped (s)')
plt.title('Drag and Mass Dependence')
plt.grid(True)
#plt.savefig('Drag and Mass Dependence')

plt.show()  
```

```python
#animated graph, from user values. It's fun but takes a long time to run, so I would recommend only uncommenting if you want to wait a while.
'''
x1 = xd
y1 = yd

x2 = xl
y2 = yl

# Frame limits
short_len = min(len(x1), len(x2))
long_len = max(len(x1), len(x2))

pause_duration = 15  # frames to simulate pause

# Extend the animation frames to include pause
total_frames = long_len + pause_duration

# Set up figure
fig, ax = plt.subplots()
ax.set_xlim(-10, max(x2) + 10)
ax.set_ylim(0, max(y2) + 2)

line1, = ax.plot([], [], 'b-', label="Line 1")
point1, = ax.plot([], [], 'bo')

line2, = ax.plot([], [], 'g-', label="Line 2")
point2, = ax.plot([], [], 'go')

# Add text for y-values next to the points
label1 = ax.text(0, 0, '', color='blue', fontsize=10)
label2 = ax.text(0, 0, '', color='green', fontsize=10)

ax.legend()

# Init function
def init():
    line1.set_data([], [])
    point1.set_data([], [])
    line2.set_data([], [])
    point2.set_data([], [])
    label1.set_text('')
    label2.set_text('')
    return line1, point1, line2, point2, label1, label2

# Update function
def update(frame):
    actual_frame = min(frame, short_len - 1)  # freeze at final frame for pause

    # Line 1: only update if still active
    if actual_frame < len(x1):
        line1.set_data(x1[:actual_frame+1], y1[:actual_frame+1])
        point1.set_data([x1[actual_frame]], [y1[actual_frame]])
        label1.set_position((x1[actual_frame], y1[actual_frame] + 0.5))
        label1.set_text(f'{y1[actual_frame]:.2f}')
    else:
        point1.set_data([], [])
        label1.set_text('')

    # Line 2: always update
    if actual_frame < len(x2):
        line2.set_data(x2[:actual_frame+1], y2[:actual_frame+1])
        point2.set_data([x2[actual_frame]], [y2[actual_frame]])
        label2.set_position((x2[actual_frame], y2[actual_frame] + 0.5))
        label2.set_text(f'{y2[actual_frame]:.2f}')
    else:
        point2.set_data([], [])
        label2.set_text('')

    return line1, point1, line2, point2, label1, label2

# Animate
ani = animation.FuncAnimation(
    fig, update,
    frames=total_frames,
    init_func=init,
    blit=True,
    interval=100,
    repeat=False
)

HTML(ani.to_jshtml())
'''
```
